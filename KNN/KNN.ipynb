```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from KNN import KNN
```

#### Ces lignes de code sont utilisées pour charger l'ensemble de données Iris, qui est un ensemble de données couramment utilisé pour la classification en apprentissage automatique. 

#### La première ligne de code définit une carte de couleurs (cmap) pour représenter les trois espèces différentes de fleurs dans l'ensemble de données Iris.

#### La deuxième ligne de code charge l'ensemble de données Iris en utilisant la fonction load_iris du module datasets de la bibliothèque sklearn. Les données sont stockées dans les tableaux X et y.




```python
cmap = ListedColormap(['#FF0000','#00FF00','#0000FF'])

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

##### Ces lignes de code créent un graphique de dispersion des données de l'ensemble de données Iris 
##### (en utilisant les fonctionnalités des pétales  X[:, 2]) et des sépales (X[:, 3]). 

```python
plt.figure()
plt.scatter(X[:,2],X[:,3], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

##### Ces lignes de code instancient un classificateur k plus proches voisins (KNN) en utilisant la classe KNN définie précédemment avec une valeur de k égale à 5 (k=5). Ensuite, les données d'entraînement (X_train et y_train) sont utilisées pour entraîner le classificateur en utilisant la méthode fit. Ensuite, le classificateur est utilisé pour faire des prédictions sur les données de test (X_test) en utilisant la méthode predict, qui renvoie un tableau de prédictions pour chaque point de données de test. 

```python
clf = KNN(k=5)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

print("Les Prédiction: " ,predictions)
```

##### La précision est calculée en divisant le nombre total de prédictions correctes par le nombre total de points de données dans l'ensemble de test (len(y_test)).

```python
acc = np.sum(predictions == y_test) / len(y_test)
print("Data accuracy",acc)
```

#### Cette ligne de code crée une nouvelle matrice de données new_data contenant une seule observation avec quatre caractéristiques. 

```python
new_data = np.array([[2, 2,2, 3]])

plt.figure()
plt.scatter(X_train[:,2], X_train[:,3], c=y_train, cmap=cmap, edgecolor='k', s=20)
plt.scatter(new_data[:,2], new_data[:,3], c='k', marker='*', s=100)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

##### clf = KNN(k=3) crée une nouvelle instance du classifieur KNN avec k = 3.
######  + clf.fit(X_train, y_train) entraîne le classifieur sur les données d'entraînement.
######  + predictions = clf.predict(X_test) prédit les classes pour les données de test.
######  + new_prediction = clf.predict(new_data) prédit la classe pour la nouvelle donnée new_data.
######  + print("Classe prédite:", new_prediction) affiche la classe prédite pour la nouvelle donnée new_data.

```python
clf = KNN(k=3)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

new_prediction = clf.predict(new_data)
print("Predicted class:",new_prediction)
```
